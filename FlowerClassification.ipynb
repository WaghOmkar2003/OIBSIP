```python
train =df.copy()
train
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(train['Species'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(train, x='SepalLengthCm', y='Species', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
train=train.drop(['Id'],axis=1)
b=(train.columns)
train.shape
```

```python
b
```

```python
train['Species'] = pd.factorize(train['Species']) [0]
train
```

```python
sns.distplot(train['Species'], hist = False, kde = True, kde_kws = {'shade': True, 'linewidth': 3})
plt.xlabel("Count")
plt.ylabel("Density")
plt.title("Density of price")
plt.legend("Price")
plt.show()
```

```python
y = train.Species
train.drop(['Species'],axis = 1,  inplace = True)
x= train
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dt = LogisticRegression()
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
import joblib
joblib.dump(dt, "IrisFlowerClassification.pkl")
```
